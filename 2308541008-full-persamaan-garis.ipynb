```python
### Grafik Persamaan Garis dengan 3 kondisi
import numpy as np
import matplotlib.pyplot as plt

def input_koordinat_08(label_08):
    x_08 = float(input(f"Nilai x dari titik {label_08} adalah "))
    y_08 = float(input(f"Nilai y dari titik {label_08} adalah "))
    return x_08, y_08

def hitung_gradien_08(x1_08, y1_08, x2_08, y2_08):
    return (y2_08 - y1_08) / (x2_08 - x1_08)

def hitung_titik_potong_y_08(gradien_08, x_08, y_08):
    return y_08 - gradien_08 * x_08

def input_gradien_dan_titik_08():
    gradien_08 = float(input("Masukkan gradien_08: "))
    x_08, y_08 = input_koordinat_08("yang dilewati")
    titik_potong_y_08 = hitung_titik_potong_y_08(gradien_08, x_08, y_08)
    return gradien_08, titik_potong_y_08

def input_persamaan_garis_08():
    a_08 = float(input("Masukkan koefisien a_08: "))
    b_08 = float(input("Masukkan koefisien b_08: "))
    c_08 = float(input("Masukkan konstanta c_08: "))
    return a_08, b_08, c_08

def main_08():
    # Kondisi 1: Diketahui dua titik
    print("Masukkan koordinat untuk garis pertama:")
    a1_08, b1_08 = input_koordinat_08("pertama dari garis pertama")
    c1_08, d1_08 = input_koordinat_08("kedua dari garis pertama")

    print("\nMasukkan koordinat untuk garis kedua:")
    e1_08, f1_08 = input_koordinat_08("pertama dari garis kedua")
    g1_08, h1_08 = input_koordinat_08("kedua dari garis kedua")

    gradien1_08 = hitung_gradien_08(a1_08, b1_08, c1_08, d1_08)
    gradien2_08 = hitung_gradien_08(e1_08, f1_08, g1_08, h1_08)
    titik_potong_y1_08 = hitung_titik_potong_y_08(gradien1_08, a1_08, b1_08)
    titik_potong_y2_08 = hitung_titik_potong_y_08(gradien2_08, e1_08, f1_08)

    # Kondisi 2: Diketahui satu gradien dan satu titik yang dilewati
    print("\nMasukkan gradien dan titik untuk garis pertama:")
    gradien3_1, titik_potong_y3_1 = input_gradien_dan_titik_08()

    print("\nMasukkan gradien dan titik untuk garis kedua:")
    gradien3_2, titik_potong_y3_2 = input_gradien_dan_titik_08()

    # Kondisi 3: Diketahui persamaan ax + by = c
    print("\nMasukkan persamaan ax + by = c untuk garis pertama:")
    a3_08, b3_08, c3_08 = input_persamaan_garis_08()
    print("Masukkan persamaan ax + by = c untuk garis kedua:")
    d3_08, e3_08, f3_08 = input_persamaan_garis_08()

    # Menghitung gradien dan intercept
    if b3_08 != 0:
        gradien4_1 = -a3_08 / b3_08
        titik_potong_y4_1 = c3_08 / b3_08
    else:
        gradien4_1 = np.inf
        titik_potong_y4_1 = c3_08 / b3_08

    if e3_08 != 0:
        gradien4_2 = -d3_08 / e3_08
        titik_potong_y4_2 = f3_08 / e3_08
    else:
        gradien4_2 = np.inf
        titik_potong_y4_2 = f3_08 / e3_08

    # Output hasil untuk ketiga kondisi
    print(f"\n[1] Persamaan garis pertama dari dua titik adalah y = {gradien1_08:.2f}x + {titik_potong_y1_08:.2f}")
    print(f"[1] Persamaan garis kedua dari dua titik adalah y = {gradien2_08:.2f}x + {titik_potong_y2_08:.2f}")

    print(f"\n[2] Persamaan garis pertama dari gradien dan titik adalah y = {gradien3_1:.2f}x + {titik_potong_y3_1:.2f}")
    print(f"[2] Persamaan garis kedua dari gradien dan titik adalah y = {gradien3_2:.2f}x + {titik_potong_y3_2:.2f}")

    print(f"\n[3] Persamaan garis pertama dari ax + by = c adalah y = {gradien4_1:.2f}x + {titik_potong_y4_1:.2f}")
    print(f"[3] Persamaan garis kedua dari ax + by = c adalah y = {gradien4_2:.2f}x + {titik_potong_y4_2:.2f}")

    # Visualisasi
    x_08 = np.linspace(-10, 10, 400)
    y1_1_08 = gradien1_08 * x_08 + titik_potong_y1_08
    y1_2_08 = gradien2_08 * x_08 + titik_potong_y2_08
    y2_1_08 = gradien3_1 * x_08 + titik_potong_y3_1
    y2_2_08 = gradien3_2 * x_08 + titik_potong_y3_2
    y3_1_08 = gradien4_1 * x_08 + titik_potong_y4_1
    y3_2_08 = gradien4_2 * x_08 + titik_potong_y4_2

    plt.figure(figsize=(10, 8))

    plt.subplot(3, 1, 1)
    plt.plot(x_08, y1_1_08, label=f'Garis pertama: y = {gradien1_08:.2f}x + {titik_potong_y1_08:.2f}', color='red')
    plt.plot(x_08, y1_2_08, label=f'Garis kedua: y = {gradien2_08:.2f}x + {titik_potong_y2_08:.2f}', color='blue')
    plt.axvline(x=0, color='grey', lw=0.5)
    plt.axhline(y=0, color='grey', lw=0.5)
    plt.legend()
    plt.title('Grafik Persamaan Garis dari Dua Titik')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(x_08, y2_1_08, label=f'Garis pertama: y = {gradien3_1:.2f}x + {titik_potong_y3_1:.2f}', color='red')
    plt.plot(x_08, y2_2_08, label=f'Garis kedua: y = {gradien3_2:.2f}x + {titik_potong_y3_2:.2f}', color='blue')
    plt.axvline(x=0, color='grey', lw=0.5)
    plt.axhline(y=0, color='grey', lw=0.5)
    plt.legend()
    plt.title('Grafik Persamaan Garis dari Gradien dan Titik')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.plot(x_08, y3_1_08, label=f'Garis pertama: y = {gradien4_1:.2f}x + {titik_potong_y4_1:.2f}', color='red')
    plt.plot(x_08, y3_2_08, label=f'Garis kedua: y = {gradien4_2:.2f}x + {titik_potong_y4_2:.2f}', color='blue')
    plt.axvline(x=0, color='grey', lw=0.5)
    plt.axhline(y=0, color='grey', lw=0.5)
    plt.legend()
    plt.title('Grafik Persamaan Garis dari ax + by = c')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main_08()
```

```python
##### Grafik Persamaan Garis
import numpy as np
import matplotlib.pyplot as plt
def input_coordinates(label):
    x_08 = float(input(f"Nilai x dari titik {label} adalah "))
    y_08 = float(input(f"Nilai y dari titik {label} adalah "))
    return x_08, y_08
print("Masukkan koordinat untuk garis pertama:")
a_08, b_08 = input_coordinates("pertama dari garis pertama")
c_08, d_08 = input_coordinates("kedua dari garis pertama")
print("\nMasukkan koordinat untuk garis kedua:")
e_08, f_08 = input_coordinates("pertama dari garis kedua")
g_08, h_08 = input_coordinates("kedua dari garis kedua")
def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)
def calculate_y_intercept(slope, x, y):
    return y - slope * x
m_1_08 = calculate_slope(a_08, b_08, c_08, d_08)
m_2_08 = calculate_slope(e_08, f_08, g_08, h_08)
c_1_08 = calculate_y_intercept(m_1_08, a_08, b_08)
c_2_08 = calculate_y_intercept(m_2_08, e_08, f_08)
print(f"\nPersamaan garis pertama adalah y = {m_1_08:.2f}x + {c_1_08:.2f}")
print(f"Persamaan garis kedua adalah y = {m_2_08:.2f}x + {c_2_08:.2f}")
if m_1_08 == m_2_08:
    if c_1_08 == c_2_08:
        print("\nKedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi.")
    else:
        print("\nKedua garis sejajar dan tidak berpotongan, sehingga tidak akan terdapat solusi.")
else:
    tp_x_08 = (c_2_08 - c_1_08) / (m_1_08 - m_2_08)
    tp_y_08 = m_1_08 * tp_x_08 + c_1_08
    print(f"\nAkan terdapat satu titik potong di ({tp_x_08:.2f}, {tp_y_08:.2f})")
x_08 = np.linspace(-10, 10, 400)
y1_08 = m_1_08 * x_08 + c_1_08
y2_08 = m_2_08 * x_08 + c_2_08
plt.plot(x_08, y1_08, label=f'Garis pertama: y = {m_1_08:.2f}x + {c_1_08:.2f}', color='red')
plt.plot(x_08, y2_08, label=f'Garis kedua: y = {m_2_08:.2f}x + {c_2_08:.2f}', color='blue')
if m_1_08 != m_2_08:
    plt.scatter(tp_x_08, tp_y_08, color='green', zorder=5)
    plt.text(tp_x_08, tp_y_08, f' ({tp_x_08:.2f}, {tp_y_08:.2f})', fontsize=12, ha='right')
plt.axvline(x=0, color='grey', lw=0.5)
plt.axhline(y=0, color='grey', lw=0.5)
plt.legend()
plt.title('Grafik Persamaan Garis')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
