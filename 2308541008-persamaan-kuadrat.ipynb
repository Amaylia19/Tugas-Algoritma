```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

def hitung_diskriminan_08(a_08, b_08, c_08):
    return b_08**2 - 4 * a_08 * c_08

def selesaikan_kuadrat_08(a_08, b_08, c_08):
    diskriminan_08 = hitung_diskriminan_08(a_08, b_08, c_08)
    x1_08 = (-b_08 + cmath.sqrt(diskriminan_08)) / (2 * a_08)
    x2_08 = (-b_08 - cmath.sqrt(diskriminan_08)) / (2 * a_08)
    return x1_08, x2_08

def cari_titik_puncak_08(a_08, b_08, c_08):
    titik_puncak_x_08 = -b_08 / (2 * a_08)
    titik_puncak_y_08 = a_08 * titik_puncak_x_08**2 + b_08 * titik_puncak_x_08 + c_08
    return titik_puncak_x_08, titik_puncak_y_08

def plot_gabungan_08(a_08, b_08, c_08, x1_08, x2_08, titik_puncak_x_08, titik_puncak_y_08):
    x_08 = np.linspace(titik_puncak_x_08 - 10, titik_puncak_x_08 + 10, 400)
    y_08 = a_08 * x_08**2 + b_08 * x_08 + c_08

    plt.figure(figsize=(10, 8))
    plt.plot(x_08, y_08, label=f'Persamaan: {a_08}x^2 + {b_08}x + {c_08} = 0')
    plt.scatter(titik_puncak_x_08, titik_puncak_y_08, color='red', zorder=5, label=f'Titik Puncak: ({titik_puncak_x_08:.2f}, {titik_puncak_y_08:.2f})')
    plt.text(titik_puncak_x_08, titik_puncak_y_08, f'({titik_puncak_x_08:.2f}, {titik_puncak_y_08:.2f})', fontsize=12, ha='right', color='red')
    
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    if x1_08.imag == 0:
        plt.axvline(x1_08.real, color='pink', linestyle='--', label=f'Akar 1: {x1_08.real:.2f}')
        plt.scatter(x1_08.real, 0, color='pink', zorder=5)
        plt.text(x1_08.real, 0, f'({x1_08.real:.2f}, 0)', fontsize=12, ha='right', color='pink')
    if x2_08.imag == 0:
        plt.axvline(x2_08.real, color='blue', linestyle='--', label=f'Akar 2: {x2_08.real:.2f}')
        plt.scatter(x2_08.real, 0, color='blue', zorder=5)
        plt.text(x2_08.real, 0, f'({x2_08.real:.2f}, 0)', fontsize=12, ha='right', color='blue')
    
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat dengan Titik Puncak dan Akar-akarnya')
    plt.show()

def bentuk_kuadrat_baru_08(akar1_08, akar2_08):
    a_08 = 1
    b_08 = -(akar1_08 + akar2_08)
    c_08 = akar1_08 * akar2_08
    return a_08, b_08, c_08

print('Program Python Mencari Akar Persamaan Kuadrat')
print('============================================')
print('\nFormat persamaan: ax^2 + bx + c = 0')

try:
    a_08 = float(input('Masukkan nilai a: '))

    if a_08 == 0:
        print("Nilai a tidak boleh 0, karena tidak bisa menjadi Persamaan kuadrat T_T")
        print("yah coba running lagi ya, yuk bisa yuk")
    else:
        print("Kurva akan membentuk keatas !" if a_08 > 0 else "Kurva akan membentuk kebawah !")
        b_08 = float(input('Masukkan nilai b: '))
        c_08 = float(input('Masukkan nilai c: '))

        print(f"Persamaan kuadrat: {a_08}x^2 + {b_08}x + {c_08} = 0")

        if c_08 == 0:
            print("Parabola akan memotong di sumbu titik (0,0)")
        else:
            print("Parabola akan memotong sumbu y pada nilai positif" if c_08 > 0 else "Parabola akan memotong sumbu y pada nilai negatif")

        diskriminan_08 = hitung_diskriminan_08(a_08, b_08, c_08)
        print(f"Nilai diskriminan: {diskriminan_08}")

        if diskriminan_08 > 0:
            print("Karena nilai Diskriminan Positif maka ada 2 kemungkinan yaitu Akar Imajiner Dan terdapat 2 solusi Akar")
        elif diskriminan_08 == 0:
            print("Diskriminan Nol, terdapat satu solusi Akar")
        else:
            print("Akar Imajiner, Dan tidak ada penyelesaian Akar")

        x1_08, x2_08 = selesaikan_kuadrat_08(a_08, b_08, c_08)
        
        print(f"Akar-akarnya adalah: {x1_08} dan {x2_08}")
        print(f"Parabola memotong sumbu y di titik (0, {c_08})")
        
        titik_puncak_x_08, titik_puncak_y_08 = cari_titik_puncak_08(a_08, b_08, c_08)
        print(f"Titik puncak adalah: ({titik_puncak_x_08}, {titik_puncak_y_08})")
        
        # Menampilkan pesan sesuai dengan posisi titik puncak
        if titik_puncak_x_08 < 0:
            print("Titik Puncak di Kiri Sumbu Y")
        elif titik_puncak_x_08 == 0:
            print("Titik Puncak di Sumbu Y")
        else:
            print("Titik Puncak di Kanan Sumbu Y")
        
        plot_gabungan_08(a_08, b_08, c_08, x1_08, x2_08, titik_puncak_x_08, titik_puncak_y_08)
except ValueError as e:
    print(e)

print("Menyusun persamaan kuadrat baru jika diketahui akar-akarnya:")
akar1_08 = float(input("Masukkan akar pertama: "))
akar2_08 = float(input("Masukkan akar kedua: "))

a_baru_08, b_baru_08, c_baru_08 = bentuk_kuadrat_baru_08(akar1_08, akar2_08)
print(f"Persamaan kuadrat baru: x^2 + ({b_baru_08})x + ({c_baru_08}) = 0")
```

```python
import cmath  
import numpy as np
import matplotlib.pyplot as plt

print('Program Python Mencari Akar Persaman Kuadrat')
print('============================================')
print()
print('Format persamaan: ax^2 + bx + c = 0')

# Menerima input dari pengguna
a_08 = int(input('Input nilai a: '))
b_08 = int(input('Input nilai b: '))
c_08 = int(input('Input nilai c: '))
print()

# Menghitung diskriminan
D_08 = (b_08 * b_08) - (4 * a_08 * c_08)
print('Diskriminan =', D_08, end='')

# Menampilkan jenis akar berdasarkan nilai diskriminan
if (D_08 > 0):
    print(' (akar real dan berbeda)')
    x1_08 = (-b_08 + cmath.sqrt(D_08)) / (2 * a_08)
    x2_08 = (-b_08 - cmath.sqrt(D_08)) / (2 * a_08)

    print('x1 =', round(x1_08.real, 2))
    print('x2 =', round(x2_08.real, 2))
elif (D_08 == 0):
    print(' (akar real dan sama)')
    x1_08 = (-b_08 + cmath.sqrt(D_08)) / (2 * a_08)
    x2_08 = x1_08

    print('x1 =', round(x1_08.real, 2))
    print('x2 =', round(x2_08.real, 2))
else:
    print(' (akar tidak real / imajiner)')
    x1_08 = (-b_08 + cmath.sqrt(D_08)) / (2 * a_08)
    x2_08 = (-b_08 - cmath.sqrt(D_08)) / (2 * a_08)
    print('x1 =', x1_08)
    print('x2 =', x2_08)

print("Mencari x1 dan x2 dalam persamaan kuadrat ax^2 + bx + c = 0 dengan rumus ABC")

# Menggunakan rumus ABC untuk menyelesaikan persamaan kuadrat
konstanta_a_08 = int(input('Masukkan variabel a: '))
konstanta_b_08 = int(input('Masukkan variabel b: '))
konstanta_c_08 = int(input('Masukkan koefisien c: '))

D_08 = (konstanta_b_08 ** 2) - (4 * konstanta_a_08 * konstanta_c_08)
print('Nilai diskriminan = ', D_08)
if D_08 < 0:
    print('Akar Imaginer')
elif D_08 == 0:
    x_08 = -konstanta_b_08 / (2 * konstanta_a_08)
    print('Akar kembar = ', x_08)
else:
    print('Akar real dan berbeda')
    x1_08 = (-konstanta_b_08 + cmath.sqrt(D_08)) / (2 * konstanta_a_08)
    x2_08 = (-konstanta_b_08 - cmath.sqrt(D_08)) / (2 * konstanta_a_08)
    print('Hasil x1 =', round(x1_08.real, 2))
    print('Hasil x2 =', round(x2_08.real, 2))

def solve_quadratic_08(a_08, b_08, c_08):
    # Menghitung diskriminan
    discriminant_08 = b_08 ** 2 - 4 * a_08 * c_08

    # Menghitung dua akar (bisa real atau kompleks)
    x1_08 = (-b_08 + cmath.sqrt(discriminant_08)) / (2 * a_08)
    x2_08 = (-b_08 - cmath.sqrt(discriminant_08)) / (2 * a_08)

    return x1_08, x2_08

# Meminta input dari pengguna
print("Masukkan koefisien untuk persamaan kuadrat (ax^2 + bx + c = 0):")
a_08 = float(input("a: "))
b_08 = float(input("b: "))
c_08 = float(input("c: "))

# Memastikan a bukan nol
if a_08 == 0:
    print("Ini bukan persamaan kuadrat karena a tidak boleh nol.")
else:
    # Menyelesaikan persamaan
    x1_08, x2_08 = solve_quadratic_08(a_08, b_08, c_08)

    # Menampilkan hasil
    print(f"Akar-akarnya adalah: {x1_08} dan {x2_08}")

    # Menampilkan kondisi parabola memotong sumbu y
    print(f"Parabola memotong sumbu y di titik (0, {c_08})")

    # Membuat data untuk grafik
    x_08 = np.linspace(-10, 10, 400)
    y_08 = a_08 * x_08 ** 2 + b_08 * x_08 + c_08

    # Membuat grafik
    plt.plot(x_08, y_08, label=f'{a_08}x^2 + {b_08}x + {c_08} = 0')
    plt.axhline(0, color='black', linewidth=0.5)
    if x1_08.imag == 0:  # Hanya plot akar real
        plt.axvline(x1_08.real, color='pink', label=f'Akar 1: {x1_08.real:.2f}')
    if x2_08.imag == 0:  # Hanya plot akar real
        plt.axvline(x2_08.real, color='blue', label=f'Akar 2: {x2_08.real:.2f}')

    # Menambahkan label dan judul
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat')
    plt.legend()
    plt.grid(True)
    plt.show()

# Meminta input dari pengguna
try:
    a_08 = float(input("Masukkan nilai a: "))
    b_08 = float(input("Masukkan nilai b: "))
    c_08 = float(input("Masukkan nilai c: "))

    if a_08 == 0:
        raise ValueError("Nilai a tidak boleh 0 untuk persamaan kuadrat.")

    # Menentukan titik puncak
    titik_puncak_x_08 = -b_08 / (2 * a_08)
    titik_puncak_y_08 = a_08 * titik_puncak_x_08 ** 2 + b_08 * titik_puncak_x_08 + c_08
    print(f"Titik puncak (vertex) adalah: ({titik_puncak_x_08}, {titik_puncak_y_08})")

    # Menampilkan kondisi parabola memotong sumbu y
    print(f"Parabola memotong sumbu y di titik (0, {c_08})")

    # Membuat range x untuk plot
    x_08 = np.linspace(titik_puncak_x_08 - 10, titik_puncak_x_08 + 10, 400)
    y_08 = a_08 * x_08 ** 2 + b_08 * x_08 + c_08

    # Membuat plot
    plt.figure(figsize=(8, 6))
    plt.plot(x_08, y_08, label=f'{a_08}x^2 + {b_08}x + {c_08}')
    plt.scatter(titik_puncak_x_08, titik_puncak_y_08, color='red', zorder=5)  # Titik puncak
    plt.text(titik_puncak_x_08, titik_puncak_y_08, f'({titik_puncak_x_08:.2f}, {titik_puncak_y_08:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()

    # Menampilkan plot
    plt.show()
except ValueError as e:
    print(e)

# Menyusun persamaan kuadrat baru jika diketahui akar-akarnya
print("Menyusun persamaan kuadrat baru jika diketahui akar-akarnya:")
akar1_08 = float(input("Masukkan akar pertama: "))
akar2_08 = float(input("Masukkan akar kedua: "))

# Menyusun persamaan kuadrat dari akar-akarnya
a_baru_08 = 1  # Koefisien untuk x^2
b_baru_08 = -(akar1_08 + akar2_08)  # Koefisien untuk x
c_baru_08 = akar1_08 * akar2_08  # Konstanta

print(f"Persamaan kuadrat baru: x^2 + ({b_baru_08})x + ({c_baru_08}) = 0")
```
