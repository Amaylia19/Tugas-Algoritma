```python
import matplotlib.pyplot as plt
import numpy as np

def hitung_08(x1_08, y1_08, x2_08, y2_08):
    garis_08 = (y2_08 - y1_08) / (x2_08 - x1_08)
    potong_08 = y1_08 - garis_08 * x1_08
    return garis_08, potong_08

def cari_tp_08(garis1_08, potong1_08, garis2_08, potong2_08):
    x_tp_08 = (potong2_08 - potong1_08) / (garis1_08 - garis2_08)
    y_tp_08 = garis1_08 * x_tp_08 + potong1_08
    return x_tp_08, y_tp_08

def gambar_08(x1_08, y1_08, x2_08, y2_08, garis1_08, potong1_08, x1_b_08, y1_b_08, x2_b_08, y2_b_08, garis2_08, potong2_08):
    x_08 = np.linspace(min(x1_08, x2_08, x1_b_08, x2_b_08) - 1, max(x1_08, x2_08, x1_b_08, x2_b_08) + 1, 100)
    y1_08 = garis1_08 * x_08 + potong1_08
    y2_08 = garis2_08 * x_08 + potong2_08

    plt.plot(x_08, y1_08, label=f'y = {garis1_08:.2f}x + {potong1_08:.2f}', color='green', linestyle='-')
    plt.scatter([x1_08, x2_08], [y1_08[0], y1_08[-1]], color='red')

    plt.plot(x_08, y2_08, label=f'y = {garis2_08:.2f}x + {potong2_08:.2f}', color='blue', linestyle='--')
    plt.scatter([x1_b_08, x2_b_08], [y2_08[0], y2_08[-1]], color='orange')

    x_tp_08, y_tp_08 = cari_tp_08(garis1_08, potong1_08, garis2_08, potong2_08)
    plt.scatter(x_tp_08, y_tp_08, color='purple', zorder=5)
    plt.text(x_tp_08, y_tp_08, f'({x_tp_08:.2f}, {y_tp_08:.2f})', fontsize=12, ha='right')

    plt.title('Persamaan Dua Garis dengan Gradient')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

    return x_tp_08, y_tp_08

def main_08():
    print("Masukkan koordinat untuk garis pertama:")
    x1_08 = float(input("Masukkan x1: "))
    y1_08 = float(input("Masukkan y1: "))
    x2_08 = float(input("Masukkan x2: "))
    y2_08 = float(input("Masukkan y2: "))

    print("Masukkan koordinat untuk garis kedua:")
    x1_b_08 = float(input("Masukkan x1: "))
    y1_b_08 = float(input("Masukkan y1: "))
    x2_b_08 = float(input("Masukkan x2: "))
    y2_b_08 = float(input("Masukkan y2: "))

    garis1_08, potong1_08 = hitung_08(x1_08, y1_08, x2_08, y2_08)
    garis2_08, potong2_08 = hitung_08(x1_b_08, y1_b_08, x2_b_08, y2_b_08)

    if garis1_08 == garis2_08:
        print("Gradient sama, tidak bisa dihitung :P")
    else:
        x_tp_08, y_tp_08 = gambar_08(
            x1_08, y1_08, x2_08, y2_08, garis1_08, potong1_08,
            x1_b_08, y1_b_08, x2_b_08, y2_b_08, garis2_08, potong2_08
        )

        print(f"Garis 1: y = {garis1_08:.2f}x + {potong1_08:.2f}")
        print(f"Garis 2: y = {garis2_08:.2f}x + {potong2_08:.2f}")
        print(f"Titik potong: ({x_tp_08:.2f}, {y_tp_08:.2f})")

        # Mencetak hasil gradient di akhir program
        print(f"Gradient Garis 1: {garis1_08:.2f}")
        print(f"Gradient Garis 2: {garis2_08:.2f}")

if __name__ == "__main__":
    main_08()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def hitung_gradien(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

def hitung_y(x, m, c):
    return m * x + c

x1_garis1_08 = float(input("Masukkan nilai x untuk titik pertama garis 1: "))
y1_garis1_08 = float(input("Masukkan nilai y untuk titik pertama garis 1: "))

x2_garis1_08 = float(input("Masukkan nilai x untuk titik kedua garis 1: "))
y2_garis1_08 = float(input("Masukkan nilai y untuk titik kedua garis 1: "))

gradien_garis1_08 = hitung_gradien(x1_garis1_08, y1_garis1_08, x2_garis1_08, y2_garis1_08)

x1_garis2_08 = float(input("Masukkan nilai x untuk titik pertama garis 2: "))
y1_garis2_08 = float(input("Masukkan nilai y untuk titik pertama garis 2: "))

x2_garis2_08 = float(input("Masukkan nilai x untuk titik kedua garis 2: "))
y2_garis2_08 = float(input("Masukkan nilai y untuk titik kedua garis 2: "))

gradien_garis2_08 = hitung_gradien(x1_garis2_08, y1_garis2_08, x2_garis2_08, y2_garis2_08)

print("Gradien garis 1:", gradien_garis1_08)
print("Gradien garis 2:", gradien_garis2_08)

# Definisikan rentang nilai x
x = np.linspace(-10, 10, 100)

# Hitung nilai y untuk kedua garis
y1 = hitung_y(x, gradien_garis1_08, y1_garis1_08)
y2 = hitung_y(x, gradien_garis2_08, y1_garis2_08)

# Plot garis pertama dan kedua
plt.plot(x, y1, label=f'Garis 1, m={gradien_garis1_08:.2f}, y1={y1_garis1_08}')
plt.plot(x, y2, label=f'Garis 2, m={gradien_garis2_08:.2f}, y1={y1_garis2_08}')

plt.xlabel('Nilai x')
plt.ylabel('Nilai y')
plt.title('Persamaan Dua Garis dan Gradient')
plt.legend()

plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
