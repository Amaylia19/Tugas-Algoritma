```python
import matplotlib.pyplot as plt
import numpy as np


def hitung_08(x1_08, y1_08, x2_08, y2_08):
   garis_08 = (y2_08 - y1_08) / (x2_08 - x1_08)
   potong_08 = y1_08 - garis_08 * x1_08
   return garis_08, potong_08


def cari_tp_08(garis1_08, potong1_08, garis2_08, potong2_08):
   x_tp_08 = (potong2_08 - potong1_08) / (garis1_08 - garis2_08)
   y_tp_08 = garis1_08 * x_tp_08 + potong1_08
   return x_tp_08,y_tp_08


def gambar_08(x1_08, y1_08, x2_08, y2_08, garis1_08, potong1_08,
                                 x1_b_08, y1_b_08, x2_b_08, y2_b_08, garis2_08, potong2_08):
   x_08 = np.linspace(min(x1_08, x2_08, x1_b_08, x2_b_08) - 1, max(x1_08, x2_08, x1_b_08, x2_b_08) + 1, 100)
   y1_08 = potong1_08 * x_08 + potong1_08
   y2_08 = garis2_08 * x_08 + potong2_08

   plt.plot(x_08, y1_08, label=f'y = {garis1_08:.2f}x + {potong1_08:.2f}', color='pink', linestyle='-')
   plt.scatter([x1_08, x2_08], [y1_08[0], y1_08[-1]], color='red')

   plt.plot(x_08, y2_08, label=f'y = {garis2_08:.2f}x + {potong2_08:.2f}', color='blue', linestyle='--')
   plt.scatter([x1_b_08, x2_b_08], [y2_08[0], y2_08[-1]], color='green')


   x_tp_08, y_tp_08 = cari_tp_08(garis1_08, potong1_08, garis2_08, potong2_08)
   plt.scatter(x_tp_08, y_tp_08, color='purple', zorder=5)
   plt.text(x_tp_08, y_tp_08, f'({x_tp_08:.2f}, {y_tp_08:.2f})', fontsize=12, ha='right')

   plt.title('Persamaan Dua Garis dengan Titik Potong')
   plt.xlabel('X')
   plt.ylabel('Y')
   plt.legend()
   plt.grid(True)
   plt.show()

   return x_tp_08, y_tp_08

def main_08():

   print("Masukkan koordinat untuk garis pertama:")
   x1_08 = float(input("Masukkan x1: "))
   y1_08 = float(input("Masukkan y1: "))
   x2_08 = float(input("Masukkan x2: "))
   y2_08 = float(input("Masukkan y2: "))


   print("Masukkan koordinat untuk garis kedua:")
   x1_b_08 = float(input("Masukkan x1: "))
   y1_b_08 = float(input("Masukkan y1: "))
   x2_b_08 = float(input("Masukkan x2: "))
   y2_b_08 = float(input("Masukkan y2: "))


   garis1_08, potong1_08 = hitung_08(x1_08, y1_08, x2_08, y2_08)
   garis2_08, potong2_08 = hitung_08(x1_b_08, y1_b_08, x2_b_08, y2_b_08)


   if garis1_08 == garis2_08:
       print("Gradient sama, gabisa di hitung :P")
   else:

       x_tp_08, y_tp_08 = gambar_08(
           x1_08, y1_08, x2_08, y2_08, garis1_08, potong1_08,
           x1_b_08, y1_b_08, x2_b_08, y2_b_08, garis2_08, potong2_08
       )

       print(f"Garis 1: y = {garis1_08:.2f}x + {potong1_08:.2f}")
       print(f"Garis 2: y = {garis2_08:.2f}x + {potong2_08:.2f}")
       print(f"Titik potong: ({x_tp_08:.2f}, {y_tp_08:.2f})")

if __name__ == "__main__":
   main_08()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def hitung_08(x1_08, y1_08, x2_08, y2_08):
   gradient_08 = (y2_08 - y1_08) / (x2_08 - x1_08)
   intercept_08 = y1_08 - gradient_08 * x1_08
   return gradient_08, intercept_08

def plot_line_08(x1_08, y1_08, x2_08, y2_08, gradient_08, intercept_08):
   x_08 = np.linspace(min(x1_08, x2_08) - 1, max(x1_08, x2_08) + 1, 100)
   y_08 = gradient_08 * x_08 + intercept_08

   plt.plot(x_08, y_08, label=f'y = {gradient_08:.2f}x + {intercept_08:.2f}', color='pink')
   plt.scatter([x1_08, x2_08], [y1_08, y2_08], color='blue')
   plt.title('Persamaan Garis dengan Gradient')
   plt.xlabel('X')
   plt.ylabel('Y')
   plt.legend()
   plt.grid(True)
   plt.show()

def main():

   x1_08 = float(input("Masukkan x1: "))
   y1_08 = float(input("Masukkan y1: "))
   x2_08 = float(input("Masukkan x2: "))
   y2_08 = float(input("Masukkan y2: "))

   gradient_08, intercept_08 = hitung_08(x1_08, y1_08, x2_08, y2_08)
   plot_line_08(x1_08, y1_08, x2_08, y2_08, gradient_08, intercept_08)

if __name__ == "__main__":
   main()
```
