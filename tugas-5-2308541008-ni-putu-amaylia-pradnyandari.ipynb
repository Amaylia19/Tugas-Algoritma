```python
n_08 = 5

segitiga_08 = [[1]]
for i in range(1, n_08):
    row_08 = [1]
    for j in range(1, i):
        row_08.append(segitiga_08[i-1][j-1] + segitiga_08[i-1][j])
    row_08.append(1)
    segitiga_08.append(row_08)

for row_08 in segitiga_08 :
    print(" ".join(str(num) for num in row_08).center(len(segitiga_08[-1])*3))
```

```python

siswa_nilai_08 = [
    ["Kharina", 80, 75, 85],
    ["Dian", 85, 90, 95],
    ["Amay", 70, 80, 75],
    ["Ratih", 95, 85, 90]
]


print("Nama\t\tRata-rata Nilai")
for siswa_08 in siswa_nilai_08 :
    nama_08 = siswa_08[0];
    rata_nilai_08 = sum(siswa_08[1:]) / len(siswa_08[1:]);
    print(f"{nama_08}\t\t{rata_nilai_08:.2f}");

print("\n")


mata_pelajaran_08 = ["Matematika", "Algoritma", "Kalkulus"];
print("Mata Pelajaran\t\tNilai Tertinggi");
for i_08, mata_08 in enumerate(mata_pelajaran_08):
    nilai_tinggi_08 = max(siswa_08[i_08+1] for siswa_08 in siswa_nilai_08);
    print(f"{mata_08}\t\t\t{nilai_tinggi_08}");
```

```python
import math

def volume_kubus_08(sisi_08):
    return sisi_08 ** 3

def volume_balok_08(panjang_08, lebar_08, tinggi_08):
    return panjang_08 * lebar_08 * tinggi_08

def volume_tabung_08(radius_08, tinggi_08):
    return math_08.pi * radius_08 ** 2 * tinggi_08

def volume_kerucut_08(radius_08, tinggi_08):
    return (1/3) * math_08.pi * radius_08 ** 2 * tinggi_08

def hitung_volume_08():
    bentuk_08 = input("Masukkan bentuk (kubus, balok, tabung, kerucut): ").lower()
    
    if bentuk_08 == "kubus":
        sisi_08 = float(input("Masukkan panjang sisi: "))
        print("Volume Kubus:", volume_kubus_08(sisi_08))
    elif bentuk_08 == "balok":
        panjang_08 = float(input("Masukkan panjang: "))
        lebar_08 = float(input("Masukkan lebar: "))
        tinggi_08 = float(input("Masukkan tinggi: "))
        print("Volume Balok:", volume_balok_08(panjang_08, lebar_08, tinggi_08))
    elif bentuk_08 == "tabung":
        radius_08 = float(input("Masukkan jari-jari alas: "))
        tinggi_08 = float(input("Masukkan tinggi: "))
        print("Volume Tabung:", volume_tabung_08(radius_08, tinggi_08))
    elif bentuk_08 == "kerucut":
        radius_08 = float(input("Masukkan jari-jari alas: "))
        tinggi_08 = float(input("Masukkan tinggi: "))
        print("Volume Kerucut:", volume_kerucut_08(radius_08, tinggi_08))
    else:
        print("Bentuk lain tidak termasuk dalam perhitungan.")

hitung_volume_08()
```
